```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
# load boston housing data set from sklearn.datasets and print it

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

```

```python
feature_name=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
```

```python
print(target)
```

```python
#transform the data set into dataframe
#data =the data we want or the independent variable also known as the x values
#target= the target variable or the price of the houses or dependent variable also known as the y values
df_x = pd.DataFrame(data, columns=feature_name)
df_y=  pd.DataFrame(target )
```

```python
#Get some statistics from the data set, count, mean
df_x.describe()
```

```python
#Initialize the linear regression model

reg = linear_model.LinearRegression()
```

```python
#split the data into 67% training and 33% testing data
x_train , x_test,y_train ,y_test= train_test_split(df_x,df_y, test_size= 0.33, random_state= 42)
```

```python
#Train the model with our training data
reg.fit(x_train, y_train)
```

```python
# print the coefficient/weight for each feature/column of our model
print(reg.coef_)
```

```python
#print the predictions on our test data
y_pred=reg.predict(x_test)
print(y_pred)
```

```python
#print the actual values
print(y_test)
```

```python
# check the model performance/ accuracy using Mean Squared Error(MSE) and numpy
print(np.mean((y_pred - y_test)**2))
```

```python
# check the model performance/ accuracy using Mean Sqraured Error(MSE) and sklearn.metrics
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```
